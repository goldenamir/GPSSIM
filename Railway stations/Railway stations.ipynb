# Converting to Excel file

```python
# It is necessary to convert it to pandas file first
# then for the conversion, I drop the last column (geometry) and specify the column names for the new df
import pandas as pd
try:
    import xlsxwriter
except:
    !pip install xlsxwriter
df1 = pd.DataFrame(db.iloc[:,:-1].values, columns = list(db.columns.values)[:-1] )

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('RailwayStations.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df1.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
